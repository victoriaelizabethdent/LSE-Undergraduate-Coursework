2. The Most Common Words in Constituency Titles

The data of most interest to me is that which shows the most commonly used words in titling a constituency. I filtered out 'North' 'South', 'East', 'West', 'Central', 'St', 'Mid', 'upon', 'and' & 'of' so I could gain a deeper insight into the titles. I also gained the average for the average amount of times a word is used in naming a constituency - including all words it was 3.13 (2 d.p.), but excluding the words listed earlier, it was 2.19. So, these words do not heavily skew the distribution, as may have been expected - as the most common words in each list 'and' and 'Glasgow', have the respective frequencies of 371 and 36.

Further analysis determines that with or without the list of words above, the median is 1.
